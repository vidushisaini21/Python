```python
import numpy as np
import pandas as pd
```

<h2>Finding Missing Data</h2>

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

```python
df.isna()
```

<h4>Formula to find missing values</h4>

```python
df.isna().sum()
```

```python
df.isna().any()
```

<h2>Removing Missing Data</h2>

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=3)  # thresh =3 means it is mandate to have 3 non null values in all rows
```

<h2>Filling the missing data</h2>

```python
df   # these are not appllied untill we write inplace = true
```

```python
df.fillna(0)  # it will fill all null values with 0
```

<h4>If we want to fill diffrent values at diffrent places of null values</h4>

```python
values = {'A':0, 'B':100, 'C':300, 'D':40}
df.fillna(value=values)
```

<h3> ### In dropna - selection is row-wise and In fillna selection is Column-wise</h3>

<h4> Fill null values with mean</h4>

```python
df
```

```python
df.fillna(df.mean())
```
