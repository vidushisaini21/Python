<h1>Series</h1> 
<h2>A series is a 1-Dimensional labeled array capable of holding any data type. The axis labels are collectively called the index.</h2>

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b', 'c']
my_list = [10,20,30]
arr = np.array([10,20,30])
dic = {6:10, 2:20,3:30}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list,index=labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(dic)
```
