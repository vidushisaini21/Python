```python
import numpy as np
```

Valid sudoku

```python
s = np.array([
    [5,3,4,6,7,8,9,1,2],
    [6,7,2,1,9,5,3,4,8],
    [1,9,8,3,4,2,5,6,7],

    [8,5,9,7,6,1,4,2,3],
    [4,2,6,8,5,3,7,9,1],
    [7,1,3,9,2,4,8,5,6],

    [9,6,1,5,3,7,2,8,4],
    [2,8,7,4,1,9,6,3,5],
    [3,4,5,2,8,6,1,7,9]   
    
    
])
```

For rows

```python
b=np.sum(s,axis = 1)  # col axis =0  row axis = 1  ( valid sudoku - every row sum = 45)
b
```

```python
b==45
```

```python
for i in b:
    if i != 45:
        print("sudoku is not valid")
        break
else:
    print("for rows it is valid")
    
```

For Columns

```python
c=np.sum(s,axis = 0)
c
```

```python
c==45
```

```python
# Columns : [Age, Math Marks, Science Marks]
data = np.array([
    [18,85,78], #student 1
    [19,92,88], #student 2
    [17,76,95], #student 3
    [18,65,85], #student 4
    [20,90,85] #student 5
])
```

```python
# Get the shape of the matrix.
data.shape
```

```python
# Find the average age of students

np.mean(data[:,0])
```

```python
# Extract maths marks of each student
data[:,1]
```

```python
# Find highest science marks
np.max(data[:,2])
```

```python
# Get details of the student who scored more than 90 in maths 
data[data[:,1] > 90]
```

```python
# Increse Math marks of all student by 5 
data[:,1]+=5
data
```

```python
# Find how many students are younger than 90
len(data[data[:,0]<19])
```

```python
# Calculate the average marks in each subjext (column-wise  mean)

np.mean(data[:,1:], axis=0)
```

```python
# Get data of students who scored at least 80 in both subjects.

data[(data[:,1]>=80) & (data[:,2]>=80)]
```

```python
# Replace all Science marks < 75 with 0.

data[:,2][data[:,2]<80]=0
data
```
