<h3>Creating a dataframe</h3>

<h3>We can create a DataFrame through list and dictionary</h3>

```python
import numpy as np 
import pandas as pd 
```

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
df
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
df2
```

```python
columns = ["Name", "Age", "City", "Salary"]
df2 = pd.DataFrame(data_list,columns=columns)
df2
```

<h3>Selection and Indexing of Columns</h3>

<h3>Creating a new Column</h3>

```python
df2["Designation"] = ["Doctor", "Engineer", "Doctor", "Teacher"]
```

```python
df2
```

<h3>Removing a column</h3>

```python
df2.drop('Designation', axis=1)
```

```python
# but it will not remove from main ddata
df2
```

<h4>Use (inplace=True) so that the deleted column will also remove from main data</h4>

```python
df2.drop('Designation', axis=1, inplace = True)
```

```python
df2
```

```python
df2.drop(["City", "Salary"], axis=1)
```

<h4>Remove rows</h4>

```python
df2.drop([0],axis=0)
```
