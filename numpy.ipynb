```python
# deep and Shallow copy
```

```python
a = np.arange(1,21)
a
```

```python
slice = a[:5]
slice
```

```python
slice = slice*10
slice
```

```python
a # a will not change by changing its any slice because it store as a different copy
```

```python
a[4:6] *10
```

```python
a # no change
```

```python
b=a
```

```python
b[0] =99
```

```python
b # b will change
```

```python
a
# a array will also chnaged (it means we are creating a deep copy when we change b then a will also change)
```

```python
# if we don't want to change we have to create a sallow copy by using (.copy())
```

```python
b= a.copy()
```

```python
b[0]=20
b
```

```python
a
```

```python
# hence b changed but a remain as it is.
# (.copy() will create a copy of a at different location but previously the location of both a nad b was same)
```

```python
# Stacking Array
```

```python
a = np.array([1,2,3,4])
b= np.array([5,6,7,8])
```

```python
 # There are three types of stack
 #    (i) Simple Stack
 #    (ii) Vertical Stack
 #    (iii) Horizontal Stack
```

```python
np.vstack((a,b))
```

```python
 # vertical stack -  .vstack((a,b))  (stack array row wise) 
```

```python
np.hstack((a,b))
```

```python
# horizontal stack -  .hstack((a,b))  (stack array column wise)
```

```python
# column stack 
np.column_stack((a,b))
```
